### Take a look into the information regarding the game

```python
pokemons.head(50)
```

```python
import seaborn as sns
# Well almost all Pokémons are assigned to one species, thus this plot is not that interesting...
species_plot = sns.catplot(
    x='species', 
    data=pokemons, 
    kind='count',
    aspect=4
).set_axis_labels('Species', '# of Pokemon');
species_plot.set_xticklabels(rotation=90)
```

To get an overview, how often specific Pokémon types are present, combine type 1 and type 2 and plot it into a barplot.

```python
pokemon_types = pokemons[['type_1']].copy()
pokemon_types = pokemon_types.rename(columns={'type_1': 'type'})
pokemon_types['type'].append(pokemons['type_2'])
# print(pokemon_types)

types_plot = sns.catplot(
    x='type', 
    data=pokemon_types, 
    kind='count',
    aspect=2,
).set_axis_labels('Species', '# of Pokemon');
types_plot.set_xticklabels(rotation=90)
types_plot.set(yticks=range(0,30, 2))
```

From this plot it can be seen, that the **most often** seen types are 
* Water
* Normal 
* Poison. 

**Often** seen are 
* grass 
* fire
* bug

Thus a Pokémon team should be able to deal with these types. Next it would be interesting, how often these types are seen in combination. 

```python
pokemon_dual_types = pokemons[pokemons['type_2'] != 'None']
dual_type_plot = sns.heatmap(
    pokemon_dual_types.groupby(['type_1', 'type_2']).size().unstack(),
    linewidths=1,
    annot=True
);
dual_type_plot.set_facecolor('gray')
#dual_type_plot.set_axis_labels('Primary type', 'Secondary type')
```

So the most common combinations that one needs to be aware of are 
* Grass/Poison
* Normal/Flying
* Bug/Poison 
* Rock/Ground
* (Rock/Water)

Some insider knowledge: Rock/Water combinations are only seen from the Pokémons that can be caught from the fossiles and I think, one has never to fight against those vs. NPCs. 

Combining the knowledge of single typed and dual typed Pokémons, it can be concluded that a team needs to deal mainly with types:
* Grass
* Normal
* Poison
* Water
* Bug
